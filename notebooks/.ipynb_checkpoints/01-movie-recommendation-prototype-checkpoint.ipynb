```python
# 2. Lấy dữ liệu phim

# Hàm lấy dữ liệu từ Supabase hoặc file local
def get_movie_data(local_file_path=None):
    """
    Lấy dữ liệu phim từ Supabase hoặc file local
    """
    # Thử lấy từ Supabase nếu có thông tin đăng nhập
    try:
        from dotenv import load_dotenv
        from supabase import create_client, Client
        
        # Load biến môi trường
        load_dotenv()
        
        # Lấy thông tin xác thực Supabase
        supabase_url = os.environ.get('SUPABASE_URL')
        supabase_key = os.environ.get('SUPABASE_KEY')
        
        if supabase_url and supabase_key:
            # Khởi tạo Supabase client
            supabase: Client = create_client(supabase_url, supabase_key)
            
            # Lấy dữ liệu từ bảng movies
            logger.info("Đang lấy dữ liệu từ Supabase...")
            response = supabase.table('movies').select('*').execute()
            
            if response and hasattr(response, 'data') and len(response.data) > 0:
                logger.info(f"Đã lấy {len(response.data)} phim từ Supabase")
                return pd.DataFrame(response.data)
            else:
                logger.warning("Không tìm thấy dữ liệu trong Supabase")
    except Exception as e:
        logger.warning(f"Không thể kết nối với Supabase: {e}")
    
    # Thử lấy từ file local
    if local_file_path:
        try:
            logger.info(f"Đang đọc dữ liệu từ file local: {local_file_path}")
            
            if local_file_path.endswith('.csv'):
                df = pd.read_csv(local_file_path)
            elif local_file_path.endswith('.json'):
                df = pd.read_json(local_file_path)
            else:
                logger.error(f"Định dạng file không được hỗ trợ: {local_file_path}")
                return None
            
            logger.info(f"Đã đọc {len(df)} phim từ file local")
            return df
        except Exception as e:
            logger.error(f"Lỗi khi đọc file local: {e}")
    
    # Nếu không có dữ liệu, tạo dữ liệu mẫu
    logger.info("Tạo dữ liệu mẫu để minh họa...")
    return create_sample_data()

# Hàm tạo dữ liệu mẫu
def create_sample_data(num_movies=50):
    """
    Tạo dữ liệu phim mẫu để minh họa
    """
    # Danh sách phim mẫu
    sample_movies = [
        {
            "id": 238,
            "title": "The Godfather",
            "overview": "Spanning the years 1945 to 1955, a chronicle of the fictional Italian-American Corleone crime family...",
            "release_date": "1972-03-14",
            "popularity": 105.761,
            "vote_average": 8.7,
            "vote_count": 18327,
            "genre_ids": [18, 80],  # Drama, Crime
            "keywords": ["mafia", "patriarch", "crime family", "organized crime", "vendetta"],
            "cast": ["Marlon Brando", "Al Pacino", "James Caan"]
        },
        {
            "id": 278,
            "title": "The Shawshank Redemption",
            "overview": "Framed in the 1940s for the double murder of his wife and her lover, upstanding banker Andy Dufresne begins a new life at the Shawshank prison...",
            "release_date": "1994-09-23",
            "popularity": 83.552,
            "vote_average": 8.7,
            "vote_count": 24451,
            "genre_ids": [18, 80],  # Drama, Crime
            "keywords": ["prison", "friendship", "corruption", "escape", "redemption"],
            "cast": ["Tim Robbins", "Morgan Freeman", "Bob Gunton"]
        },
        {
            "id": 240,
            "title": "The Godfather Part II",
            "overview": "In the continuing saga of the Corleone crime family, a young Vito Corleone grows up in Sicily and in 1910s New York...",
            "release_date": "1974-12-20",
            "popularity": 65.213,
            "vote_average": 8.5,
            "vote_count": 10816,
            "genre_ids": [18, 80],  # Drama, Crime
            "keywords": ["mafia", "rise to power", "sequel", "organized crime", "family"],
            "cast": ["Al Pacino", "Robert De Niro", "Robert Duvall"]
        },
        {
            "id": 424,
            "title": "Schindler's List",
            "overview": "The true story of how businessman Oskar Schindler saved over a thousand Jewish lives from the Nazis...",
            "release_date": "1993-12-15",
            "popularity": 55.143,
            "vote_average": 8.6,
            "vote_count": 14075,
            "genre_ids": [18, 36, 10752],  # Drama, History, War
            "keywords": ["holocaust", "jew", "world war ii", "nazi", "genocide"],
            "cast": ["Liam Neeson", "Ben Kingsley", "Ralph Fiennes"]
        },
        {
            "id": 129,
            "title": "Spirited Away",
            "overview": "A young girl, Chihiro, becomes trapped in a strange new world of spirits. When her parents undergo a mysterious transformation...",
            "release_date": "2001-07-20",
            "popularity": 73.933,
            "vote_average": 8.5,
            "vote_count": 14432,
            "genre_ids": [16, 10751, 14],  # Animation, Family, Fantasy
            "keywords": ["spirit", "anime", "alternative world", "coming of age", "surrealism"],
            "cast": ["Rumi Hiiragi", "Miyu Irino", "Mari Natsuki"]
        }
    ]
    
    # Mở rộng danh sách phim mẫu
    if num_movies > len(sample_movies):
        # Tạo thêm phim mẫu
        genres = [
            [28], [12], [16], [35], [80], [99], [18], [10751], [14], [36], 
            [27], [10402], [9648], [10749], [878], [53], [10752], [37]
        ]
        keywords_list = [
            ["action", "explosion", "fight"], 
            ["adventure", "journey", "quest"],
            ["comedy", "funny", "humor"],
            ["romance", "love", "relationship"],
            ["thriller", "suspense", "mystery"],
            ["horror", "scary", "fear"],
            ["science fiction", "space", "future"],
            ["documentary", "true story", "history"]
        ]
        
        for i in range(len(sample_movies), num_movies):
            # Tạo phim ngẫu nhiên
            import random
            random_genre = random.choice(genres)
            random_keywords = random.choice(keywords_list)
            
            new_movie = {
                "id": 1000 + i,
                "title": f"Sample Movie {i+1}",
                "overview": f"This is a sample movie overview for movie {i+1}.",
                "release_date": f"{random.randint(1970, 2023)}-{random.randint(1, 12)}-{random.randint(1, 28)}",
                "popularity": round(random.uniform(10, 100), 3),
                "vote_average": round(random.uniform(3, 9), 1),
                "vote_count": random.randint(100, 20000),
                "genre_ids": random_genre,
                "keywords": random_keywords,
                "cast": [f"Actor {j+1}" for j in range(3)]
            }
            
            sample_movies.append(new_movie)
    
    return pd.DataFrame(sample_movies)

# Lấy dữ liệu phim
# Thử tìm kiếm file dữ liệu phim ở các vị trí phổ biến
possible_data_paths = [
    '../data/movies.csv',
    '../data/movies.json',
    './data/movies.csv',
    './data/movies.json'
]

# Tìm file dữ liệu đầu tiên tồn tại
data_path = None
for path in possible_data_paths:
    if os.path.exists(path):
        data_path = path
        break

# Lấy dữ liệu
movies_df = get_movie_data(data_path)

# Hiển thị thông tin dữ liệu
print(f"Số lượng phim: {len(movies_df)}")
movies_df.head()
```

```python
# 3. Tiền xử lý dữ liệu

# Hàm tiền xử lý dữ liệu
def preprocess_data(df):
    """
    Tiền xử lý dữ liệu phim
    """
    # Tạo bản sao để không ảnh hưởng đến dữ liệu gốc
    processed_df = df.copy()
    
    # Kiểm tra và đảm bảo có các cột cần thiết
    required_columns = ['id', 'title', 'overview']
    for col in required_columns:
        if col not in processed_df.columns:
            logger.warning(f"Không tìm thấy cột {col}, thêm cột trống")
            processed_df[col] = ""
    
    # Xử lý giá trị null
    processed_df['overview'] = processed_df['overview'].fillna('')
    
    # Chuyển đổi release_date thành datetime nếu tồn tại
    if 'release_date' in processed_df.columns:
        processed_df['release_date'] = pd.to_datetime(processed_df['release_date'], errors='coerce')
        # Trích xuất năm phát hành
        processed_df['release_year'] = processed_df['release_date'].dt.year
    
    # Kiểm tra dữ liệu genre_ids, keywords, cast
    list_columns = ['genre_ids', 'keywords', 'cast']
    for col in list_columns:
        if col in processed_df.columns:
            # Chuyển đổi giá trị None thành list rỗng
            processed_df[col] = processed_df[col].apply(lambda x: [] if x is None else x)
            
            # Kiểm tra nếu là list chuỗi JSON, chuyển đổi thành list Python
            if processed_df[col].dtype == 'object':
                try:
                    processed_df[col] = processed_df[col].apply(
                        lambda x: json.loads(x) if isinstance(x, str) and x.startswith('[') else x
                    )
                except Exception as e:
                    logger.warning(f"Không thể chuyển đổi cột {col} thành list: {e}")
    
    return processed_df

# Tiền xử lý dữ liệu
processed_movies_df = preprocess_data(movies_df)

# Hiển thị thông tin dữ liệu sau khi tiền xử lý
print("Thông tin dữ liệu sau khi tiền xử lý:")
processed_movies_df.info()

# Hiển thị mẫu dữ liệu
processed_movies_df.head()
```

```python
# 4. Tạo embedding với Sentence Transformers

# Kiểm tra xem Sentence Transformers có sẵn không
if not SENTENCE_TRANSFORMERS_AVAILABLE:
    raise ImportError("Sentence Transformers không được cài đặt. Hãy chạy: pip install sentence-transformers")

# Tạo văn bản kết hợp cho mỗi phim
def create_movie_texts(movies_df):
    """
    Tạo văn bản mô tả phim để tạo embedding
    """
    movie_texts = []
    
    for _, movie in movies_df.iterrows():
        # Bắt đầu với tiêu đề
        text = f"{movie['title']}. "
        
        # Thêm tóm tắt nội dung
        if 'overview' in movie and pd.notna(movie['overview']) and movie['overview'] != '':
            text += movie['overview'] + " "
            
        # Thêm thể loại nếu có
        if 'genre_ids' in movie and isinstance(movie['genre_ids'], list) and movie['genre_ids']:
            genre_mapping = {
                28: "Action", 12: "Adventure", 16: "Animation", 35: "Comedy", 80: "Crime",
                99: "Documentary", 18: "Drama", 10751: "Family", 14: "Fantasy", 36: "History",
                27: "Horror", 10402: "Music", 9648: "Mystery", 10749: "Romance", 878: "Science Fiction",
                10770: "TV Movie", 53: "Thriller", 10752: "War", 37: "Western"
            }
            
            genres = [genre_mapping.get(gid, f"Genre_{gid}") for gid in movie['genre_ids']]
            text += "Thể loại: " + ", ".join(genres) + ". "
        
        # Thêm từ khóa nếu có
        if 'keywords' in movie and isinstance(movie['keywords'], list) and movie['keywords']:
            text += "Từ khóa: " + ", ".join(movie['keywords']) + ". "
            
        # Thêm diễn viên nếu có
        if 'cast' in movie and isinstance(movie['cast'], list) and movie['cast']:
            text += "Diễn viên: " + ", ".join(movie['cast']) + "."
            
        movie_texts.append(text)
    
    return movie_texts

# Tạo văn bản mô tả phim
movie_texts = create_movie_texts(processed_movies_df)

# Hiển thị một số ví dụ văn bản mô tả phim
print("Ví dụ văn bản mô tả phim:")
for i, text in enumerate(movie_texts[:3]):
    print(f"{i+1}. {text}")
    print("-" * 100)

# Tải mô hình Sentence Transformer
model_name = "sentence-transformers/all-MiniLM-L6-v2"
print(f"Đang tải mô hình Sentence Transformer: {model_name}")
model = SentenceTransformer(model_name)

# Tạo embedding
print("Đang tạo embedding cho các phim...")
start_time = time.time()
embeddings = model.encode(movie_texts, show_progress_bar=True)
end_time = time.time()

print(f"Đã tạo {len(embeddings)} embedding vectors, mỗi vector có kích thước {embeddings.shape[1]}")
print(f"Thời gian tạo embedding: {end_time - start_time:.2f} giây")

# Lưu embedding và dữ liệu để sử dụng lại nếu cần
embeddings_path = '../models/movie_embeddings.joblib'
Path('../models').mkdir(exist_ok=True)
joblib.dump({
    'embeddings': embeddings,
    'movies_df': processed_movies_df,
    'movie_texts': movie_texts
}, embeddings_path)
print(f"Đã lưu embedding vào: {embeddings_path}")
```

```python
# 5. Xây dựng FAISS index

# Kiểm tra xem FAISS có sẵn không
if not FAISS_AVAILABLE:
    raise ImportError("FAISS không được cài đặt. Hãy chạy: pip install faiss-cpu")

# Tạo FAISS index để tìm kiếm vector hiệu quả
def create_faiss_index(embeddings, index_type='flat'):
    """
    Tạo FAISS index cho tìm kiếm vector hiệu quả
    
    Args:
        embeddings: Mảng embeddings
        index_type: Loại index ('flat', 'ivf', 'hnsw')
    
    Returns:
        FAISS index
    """
    # Lấy kích thước vector
    vector_dim = embeddings.shape[1]
    
    # Chuyển embeddings sang float32 (yêu cầu của FAISS)
    embeddings_32 = embeddings.astype('float32')
    
    # Chuẩn hóa embeddings để sử dụng cosine similarity
    faiss.normalize_L2(embeddings_32)
    
    # Tạo index dựa trên loại
    if index_type == 'flat':
        # Index chính xác nhất, nhưng chậm với dữ liệu lớn
        index = faiss.IndexFlatIP(vector_dim)  # IP = Inner Product
    elif index_type == 'ivf':
        # Index nhanh hơn, sử dụng clustering
        n_clusters = min(int(embeddings.shape[0] / 10), 100)  # Số lượng cluster
        quantizer = faiss.IndexFlatIP(vector_dim)
        index = faiss.IndexIVFFlat(quantizer, vector_dim, n_clusters, faiss.METRIC_INNER_PRODUCT)
        # Cần train index trước khi sử dụng
        index.train(embeddings_32)
    elif index_type == 'hnsw':
        # Hierarchical Navigable Small World - hiệu quả cho tập dữ liệu lớn
        index = faiss.IndexHNSWFlat(vector_dim, 32, faiss.METRIC_INNER_PRODUCT)  # 32 là số kết nối mỗi node
    else:
        raise ValueError(f"Loại index không hợp lệ: {index_type}")
    
    # Thêm embeddings vào index
    index.add(embeddings_32)
    
    print(f"Đã tạo FAISS index loại {index_type} với {index.ntotal} vectors")
    return index

# Đo thời gian tạo các loại index khác nhau
index_types = ['flat', 'ivf', 'hnsw']
indices = {}
times = {}

for index_type in index_types:
    print(f"\nĐang tạo FAISS index loại {index_type}...")
    start_time = time.time()
    try:
        indices[index_type] = create_faiss_index(embeddings, index_type)
        end_time = time.time()
        times[index_type] = end_time - start_time
        print(f"Thời gian tạo index {index_type}: {times[index_type]:.2f} giây")
    except Exception as e:
        print(f"Lỗi khi tạo index {index_type}: {e}")

# Lưu index tốt nhất (flat cho độ chính xác, hnsw cho tốc độ với dữ liệu lớn)
best_index_type = 'flat' if len(embeddings) < 10000 else 'hnsw'
if best_index_type in indices:
    index_path = f'../models/faiss_index_{best_index_type}.bin'
    faiss.write_index(indices[best_index_type], index_path)
    print(f"Đã lưu FAISS index loại {best_index_type} vào: {index_path}")

# So sánh hiệu suất các loại index
plt.figure(figsize=(10, 6))
plt.bar(times.keys(), times.values())
plt.title('Thời gian tạo FAISS index theo loại')
plt.xlabel('Loại index')
plt.ylabel('Thời gian (giây)')
plt.show()
```

```python
# 6. Thực hiện gợi ý phim

# Lớp gợi ý phim sử dụng FAISS
class MovieRecommender:
    """Hệ thống gợi ý phim sử dụng Sentence Transformers và FAISS"""
    
    def __init__(self, movies_df, embeddings, faiss_index, model=None):
        """
        Khởi tạo hệ thống gợi ý
        
        Args:
            movies_df: DataFrame chứa thông tin phim
            embeddings: Mảng embeddings của phim
            faiss_index: FAISS index để tìm kiếm
            model: Mô hình Sentence Transformer (tùy chọn, cần cho tìm kiếm theo văn bản)
        """
        self.movies_df = movies_df
        self.embeddings = embeddings
        self.faiss_index = faiss_index
        self.model = model
        
        # Tạo ánh xạ từ title sang index
        self.title_to_idx = {title: idx for idx, title in enumerate(self.movies_df['title'])}
    
    def recommend_by_movie_id(self, movie_id, k=5):
        """
        Gợi ý phim dựa trên ID
        
        Args:
            movie_id: ID của phim
            k: Số lượng phim gợi ý
            
        Returns:
            DataFrame chứa phim gợi ý và điểm tương đồng
        """
        # Tìm index của phim
        movie_idx = self.movies_df[self.movies_df['id'] == movie_id].index
        if len(movie_idx) == 0:
            print(f"Không tìm thấy phim với ID: {movie_id}")
            return pd.DataFrame()
        
        movie_idx = movie_idx[0]
        return self.recommend_by_idx(movie_idx, k)
    
    def recommend_by_movie_title(self, title, k=5):
        """
        Gợi ý phim dựa trên tiêu đề
        
        Args:
            title: Tiêu đề phim
            k: Số lượng phim gợi ý
            
        Returns:
            DataFrame chứa phim gợi ý và điểm tương đồng
        """
        # Tìm index của phim
        if title in self.title_to_idx:
            movie_idx = self.title_to_idx[title]
            return self.recommend_by_idx(movie_idx, k)
        else:
            # Tìm phim có tiêu đề gần nhất
            closest_titles = sorted(self.title_to_idx.keys(), 
                                  key=lambda x: self._levenshtein_distance(x.lower(), title.lower()))
            if closest_titles:
                print(f"Không tìm thấy phim '{title}'. Có phải bạn muốn tìm '{closest_titles[0]}'?")
                return self.recommend_by_idx(self.title_to_idx[closest_titles[0]], k)
            else:
                print(f"Không tìm thấy phim với tiêu đề: {title}")
                return pd.DataFrame()
    
    def recommend_by_idx(self, idx, k=5):
        """
        Gợi ý phim dựa trên index
        
        Args:
            idx: Index của phim trong DataFrame
            k: Số lượng phim gợi ý
            
        Returns:
            DataFrame chứa phim gợi ý và điểm tương đồng
        """
        # Lấy embedding của phim
        query_vector = self.embeddings[idx].reshape(1, -1).astype('float32')
        
        # Chuẩn hóa vector truy vấn
        faiss.normalize_L2(query_vector)
        
        # Tìm kiếm phim tương tự
        k_results = min(k + 1, self.faiss_index.ntotal)  # +1 vì sẽ loại bỏ phim gốc
        distances, indices = self.faiss_index.search(query_vector, k_results)
        
        # Chuyển đổi kết quả
        distances = distances[0]
        indices = indices[0]
        
        # Loại bỏ phim gốc nếu nó nằm trong kết quả
        mask = indices != idx
        distances = distances[mask][:k]
        indices = indices[mask][:k]
        
        # Tạo DataFrame kết quả
        results = self.movies_df.iloc[indices].copy()
        results['similarity_score'] = distances
        
        return results
    
    def recommend_by_text(self, query_text, k=5):
        """
        Gợi ý phim dựa trên văn bản truy vấn
        
        Args:
            query_text: Văn bản truy vấn
            k: Số lượng phim gợi ý
            
        Returns:
            DataFrame chứa phim gợi ý và điểm tương đồng
        """
        if self.model is None:
            print("Cần có mô hình Sentence Transformer để tìm kiếm theo văn bản")
            return pd.DataFrame()
        
        # Tạo embedding cho văn bản truy vấn
        query_embedding = self.model.encode(query_text)
        query_embedding = query_embedding.reshape(1, -1).astype('float32')
        
        # Chuẩn hóa embedding
        faiss.normalize_L2(query_embedding)
        
        # Tìm kiếm phim tương tự
        distances, indices = self.faiss_index.search(query_embedding, k)
        
        # Chuyển đổi kết quả
        distances = distances[0]
        indices = indices[0]
        
        # Tạo DataFrame kết quả
        results = self.movies_df.iloc[indices].copy()
        results['similarity_score'] = distances
        
        return results
    
    def _levenshtein_distance(self, s1, s2):
        """Tính khoảng cách Levenshtein giữa hai chuỗi"""
        if len(s1) < len(s2):
            return self._levenshtein_distance(s2, s1)
        
        if len(s2) == 0:
            return len(s1)
        
        previous_row = range(len(s2) + 1)
        for i, c1 in enumerate(s1):
            current_row = [i + 1]
            for j, c2 in enumerate(s2):
                insertions = previous_row[j + 1] + 1
                deletions = current_row[j] + 1
                substitutions = previous_row[j] + (c1 != c2)
                current_row.append(min(insertions, deletions, substitutions))
            previous_row = current_row
            
        return previous_row[-1]

# Tạo hệ thống gợi ý
best_index_type = 'flat' if len(embeddings) < 10000 else 'hnsw'
recommender = MovieRecommender(
    movies_df=processed_movies_df,
    embeddings=embeddings,
    faiss_index=indices[best_index_type],
    model=model
)

# Hiển thị một số phim mẫu để thử nghiệm
print("Các phim mẫu để thử nghiệm:")
sample_movies = processed_movies_df.head(5)
for i, row in enumerate(sample_movies.iterrows()):
    _, movie = row
    print(f"{i+1}. {movie['title']} (ID: {movie['id']})")

# Thử nghiệm gợi ý dựa trên phim
if len(processed_movies_df) > 0:
    test_movie = processed_movies_df.iloc[0]
    print(f"\nGợi ý phim tương tự với '{test_movie['title']}':")
    recommendations = recommender.recommend_by_movie_title(test_movie['title'])
    
    # Hiển thị kết quả
    for i, (_, movie) in enumerate(recommendations.iterrows()):
        print(f"{i+1}. {movie['title']} (Độ tương đồng: {movie['similarity_score']:.4f})")
        if 'overview' in movie and pd.notna(movie['overview']) and movie['overview']:
            overview = movie['overview']
            if len(overview) > 100:
                overview = overview[:100] + "..."
            print(f"   Tóm tắt: {overview}")
        print()

# Thử nghiệm gợi ý dựa trên văn bản
test_queries = [
    "drama",
    "ghost",
    "cartoon and family",
    "mafia"
]

for query in test_queries:
    print(f"\nGợi ý phim cho truy vấn: '{query}'")
    recommendations = recommender.recommend_by_text(query)
    
    # Hiển thị kết quả
    for i, (_, movie) in enumerate(recommendations.iterrows()):
        print(f"{i+1}. {movie['title']} (Độ tương đồng: {movie['similarity_score']:.4f})")
    print()
```

```python
# 7. So sánh hiệu suất

# So sánh hiệu suất tìm kiếm giữa các phương pháp
def compare_search_performance(n_queries=20, k=5):
    """
    So sánh hiệu suất giữa tìm kiếm cosine similarity thông thường và FAISS
    
    Args:
        n_queries: Số lượng truy vấn để kiểm tra
        k: Số lượng phim gợi ý mỗi lần
    """
    # Lấy mẫu ngẫu nhiên các phim để thử nghiệm
    import random
    sample_indices = random.sample(range(len(processed_movies_df)), min(n_queries, len(processed_movies_df)))
    
    # Tạo mảng embeddings dưới dạng float32 cho tính toán cosine similarity
    embeddings_32 = embeddings.astype('float32')
    
    # Chuẩn bị vector đầu vào cho các truy vấn
    query_vectors = [embeddings_32[idx].reshape(1, -1) for idx in sample_indices]
    
    # Đo thời gian tìm kiếm với cosine similarity
    start_time = time.time()
    cosine_results = []
    for idx, query in zip(sample_indices, query_vectors):
        # Tính toán cosine similarity với tất cả embeddings
        similarities = cosine_similarity(query, embeddings_32)[0]
        
        # Lấy k phim tương tự nhất (trừ chính nó)
        similar_indices = similarities.argsort()[::-1]
        # Loại bỏ phim gốc
        similar_indices = similar_indices[similar_indices != idx][:k]
        
        cosine_results.append(similar_indices)
    cosine_time = time.time() - start_time
    
    # Đo thời gian tìm kiếm với FAISS
    start_time = time.time()
    faiss_results = []
    for idx, query in zip(sample_indices, query_vectors):
        # Tìm kiếm với FAISS
        query_normalized = query.copy()
        faiss.normalize_L2(query_normalized)
        _, indices = recommender.faiss_index.search(query_normalized, k + 1)
        
        # Lấy kết quả (loại bỏ phim gốc)
        result_indices = indices[0]
        result_indices = result_indices[result_indices != idx][:k]
        
        faiss_results.append(result_indices)
    faiss_time = time.time() - start_time
    
    # Tính độ trùng khớp giữa hai phương pháp
    overlap = 0
    total_items = 0
    for cosine_idx, faiss_idx in zip(cosine_results, faiss_results):
        common_items = set(cosine_idx) & set(faiss_idx)
        overlap += len(common_items)
        total_items += min(len(cosine_idx), len(faiss_idx))
    
    overlap_percentage = (overlap / total_items) * 100 if total_items > 0 else 0
    
    # In kết quả
    print(f"So sánh hiệu suất tìm kiếm với {n_queries} truy vấn, mỗi truy vấn trả về {k} kết quả:")
    print(f"1. Cosine similarity: {cosine_time:.4f} giây ({cosine_time/n_queries*1000:.2f} ms/truy vấn)")
    print(f"2. FAISS: {faiss_time:.4f} giây ({faiss_time/n_queries*1000:.2f} ms/truy vấn)")
    print(f"Tăng tốc: {cosine_time/faiss_time:.2f}x")
    print(f"Độ trùng khớp kết quả: {overlap_percentage:.2f}%")
    
    # Vẽ biểu đồ so sánh
    plt.figure(figsize=(10, 6))
    methods = ['Cosine Similarity', 'FAISS']
    times = [cosine_time, faiss_time]
    plt.bar(methods, times)
    plt.title('Thời gian tìm kiếm (giây)')
    plt.ylabel('Thời gian (giây)')
    
    # Thêm nhãn thời gian
    for i, time_val in enumerate(times):
        plt.text(i, time_val + 0.05, f"{time_val:.4f}s", ha='center')
    
    plt.show()
    
    # Vẽ biểu đồ thời gian trung bình mỗi truy vấn
    plt.figure(figsize=(10, 6))
    avg_times = [cosine_time/n_queries*1000, faiss_time/n_queries*1000]  # Đổi sang millisecond
    plt.bar(methods, avg_times)
    plt.title('Thời gian trung bình mỗi truy vấn (ms)')
    plt.ylabel('Thời gian (ms)')
    
    # Thêm nhãn thời gian
    for i, time_val in enumerate(avg_times):
        plt.text(i, time_val + 1, f"{time_val:.2f}ms", ha='center')
    
    plt.show()

# Thực hiện so sánh hiệu suất
compare_search_performance(20)
```
